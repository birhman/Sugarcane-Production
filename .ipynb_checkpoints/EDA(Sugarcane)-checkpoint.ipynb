# Importing Libraries and Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sugar = pd.read_csv('List of Countries by Sugarcane Production.csv')
```

# Data Cleaning

```python
sugar.head()
```

```python
sugar.shape
```

```python
sugar['Production (Tons)'] =  sugar['Production (Tons)'].str.replace(".", "") 
sugar['Production per Person (Kg)'] = sugar['Production per Person (Kg)'].str.replace("." , "").str.replace("," , "")
sugar["Acreage (Hectare)"] = sugar['Acreage (Hectare)'].str.replace("." , "")
sugar['Yield (Kg / Hectare)'] = sugar ['Yield (Kg / Hectare)'].str.replace(".", "").str.replace("," , ".")
```

```python
sugar.rename(columns = {'Production (Tons)' : 'Production(Tons)'}, inplace = True)
sugar.rename(columns = {'Production per Person (Kg)' : 'Production_per_Person(Kg)'}, inplace = True)
sugar.rename(columns = {'Acreage (Hectare)' : 'Acreage(Hectare)'}, inplace = True)
sugar.rename(columns = {'Yield (Kg / Hectare)' : 'Yield(Kg / Hectare'}, inplace = True)
```

```python
sugar.isna().sum()
```

```python
sugar[sugar['Acreage(Hectare)'].isnull()]
sugar[sugar['Yield(Kg / Hectare'].isnull()]
```

```python
sugar = sugar.dropna().reset_index()
```

```python
sugar.head()
```

```python
sugar.isna().sum()
```

```python
sugar = sugar.drop(sugar.iloc[:, 0:2], axis=1)
```

```python
sugar.head()
```

```python
for i in sugar.iloc[:, 2:6]:
    sugar[i] = sugar[i].astype(float)
```

```python
sugar.dtypes
```

```python
sugar.nunique()
```

# BiVariate Analysis

```python
sugar.head(1)
```

## Which Country produces most Sugercane?

```python
sugar_new = sugar[['Country', 'Production(Tons)']].set_index('Country')
```

```python
sugar_new.head()
```

```python
sugar_new['Production(Tons)_percentage'] = round(sugar_new['Production(Tons)'] * 100 / sugar_new['Production(Tons)'].sum(), 2)
```

```python
sugar_new
```

```python
sugar_new['Production(Tons)_percentage'].head(5).plot(kind = 'pie', autopct = '%.2f')
plt.show()
```

```python
sugar_prod = sugar.sort_values('Production(Tons)', ascending = False)
plt.title('Country wise Production')
ax = sns.barplot(data = sugar_prod.head(10), x = 'Country', y = 'Production(Tons)', hue = 'Country')
plt.xticks(rotation = 90)
plt.show()
```

## Which country has highest land for Sugercane Production? 

```python
sugar_acr = sugar.sort_values('Acreage(Hectare)', ascending = False)
plt.title('Land for Sugarcane Cultivation')
sns.barplot(data = sugar_acr.head(10), x = 'Country', y = 'Acreage(Hectare)', hue = 'Country' )
plt.xticks(rotation = 90)
plt.show()
```

## Which country has highest Yield

```python
plt.title('Country wise Yield (Kg/Hectare)')
sugar_yld = sugar.sort_values('Yield(Kg / Hectare', ascending = False)
sns.barplot(data = sugar_yld.head(10), x = 'Country', y = 'Yield(Kg / Hectare', hue = 'Country')
plt.xticks(rotation = 90)
plt.show()
```

## Countries with highest production per capita

## Analysis for continents 

```python
sugar_continent = sugar.groupby('Continent').sum(numeric_only=True)
```

```python
plt.title('Continents to grow Sugarcane')
sns.barplot(data = sugar_continent.sort_values(by = 'Production(Tons)' ,ascending=False), x = 'Continent', y = 'Production(Tons)', hue = 'Continent')
plt.show()
```

```python
sugar_continent['no_of_countries'] = sugar.groupby('Continent').count()['Country']
```

```python
sugar_continent
```

```python
continent_names = sugar_continent.index.to_list()
plt.title('Continents vs Cultivated Area')
sns.lineplot(data = sugar_continent, x = 'no_of_countries', y = 'Production(Tons)')
plt.xticks(sugar_continent['no_of_countries'], continent_names, rotation = 90)
plt.show()
```

```python
plt.pie(sugar_continent['Production(Tons)'], autopct='%.2f', labels = continent_names)
plt.show()
```
